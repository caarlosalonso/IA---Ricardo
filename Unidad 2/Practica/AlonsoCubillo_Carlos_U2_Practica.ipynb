## Ejercicio 1.1: Análisis de Código con Refinamiento

**Objetivo:** Crear un prompt para analizar código Python que mejore iterativamente.

### Código a analizar:

````python
def procesar_datos(datos):
    resultado = []
    for i in range(len(datos)):
        if datos[i] != None:
            if type(datos[i]) == str:
                resultado.append(datos[i].strip().lower())
            else:
                resultado.append(datos[i])
    return resultado

def buscar(lista, elemento):
    for i in range(len(lista)):
        if lista[i] == elemento:
            return i
    return -1
````

### Iteración 1 - Prompt básico

- Escribe un prompt simple para analizar el código
- Ejecútalo y documenta la respuesta
- Identifica qué falta o qué podría mejorar

#### Prompt básico (Iteración 1)

```
Analiza este código de Python, dime que hace, y si falta o si se puede mejorar algo
```

#### Respuesta del modelo (Iteración 1)

El código tiene dos funciones: una limpia una lista eliminando valores `None` y normalizando textos (quitando espacios y pasando a minúsculas), y otra busca un elemento devolviendo su índice o `-1` si no lo encuentra.

Funciona correctamente, pero se puede mejorar en estilo y eficiencia: usa patrones poco idiomáticos en Python (como recorrer con índices), hace comparaciones mejorables (`None` y tipos), y reimplementa funcionalidades que ya existen en el lenguaje. También puede simplificarse para ser más claro, compacto y mantenible.

def procesar_datos(datos):
    """Procesa una lista eliminando valores None y normalizando strings."""
    resultado = []
    
    for item in datos:
        if item is not None:
            if isinstance(item, str):
                resultado.append(item.strip().lower())
            else:
                resultado.append(item)
    
    return resultado


def buscar(lista, elemento):

## Ejercicio 1.3: Razonamiento con Chain of Thought

**Objetivo:** Aplicar CoT para resolver problemas de razonamiento.

### Problema:

Una empresa de software tiene 3 equipos:
- **Equipo Frontend:** 4 desarrolladores, cada uno puede completar 2 features/semana
- **Equipo Backend:** 3 desarrolladores, cada uno puede completar 1.5 features/semana
- **Equipo QA:** 2 testers, cada uno puede validar 5 features/semana

Para el próximo release se necesitan **40 features** desarrolladas y validadas.

Considerando que QA solo puede validar features ya completadas:
1. ¿Cuántas semanas mínimo se necesitan?
2. ¿Hay algún cuello de botella? ¿Cuál?

### Sin Chain of Thought

#### Prompt sin CoT

```
Una empresa tiene:
- Equipo Frontend: 4 desarrolladores, 2 features/semana cada uno
- Equipo Backend: 3 desarrolladores, 1.5 features/semana cada uno
- Equipo QA: 2 testers, 5 features/semana cada uno

Se necesitan 40 features desarrolladas y validadas. QA solo valida features ya completadas.
¿Cuántas semanas mínimo se necesitan? ¿Hay cuello de botella?
```

#### Respuesta del modelo (sin CoT)

Frontend produce 8 features/semana, Backend 4.5 features/semana y QA valida 10 features/semana.
Como QA solo valida lo ya desarrollado, la velocidad total la limita el equipo más lento del desarrollo: Backend (4.5/semana).

Para 40 features:
40 ÷ 4.5 = 8.89 semanas → se redondea a 9 semanas mínimo.

Respuesta: 9 semanas mínimo. Sí, el cuello de botella está en Backend.

### Con Chain of Thought estructurado (pasos explícitos)

#### Prompt con CoT

```
Resuelve el siguiente problema de planificación de equipos paso a paso.
Muestra todos los cálculos intermedios claramente antes de dar la respuesta final.

Datos:
- Equipo Frontend: 4 desarrolladores, cada uno completa 2 features/semana
- Equipo Backend: 3 desarrolladores, cada uno completa 1.5 features/semana
- Equipo QA: 2 testers, cada uno valida 5 features/semana
- Meta: 40 features desarrolladas Y validadas
- Restricción: QA solo puede validar features ya completadas por desarrollo

Sigue estos pasos:
Paso 1: Calcula la capacidad semanal total de desarrollo (Frontend + Backend).
Paso 2: Calcula la capacidad semanal total de QA.
Paso 3: Determina cuántas semanas necesita desarrollo para completar 40 features.
Paso 4: Determina cuántas semanas necesita QA para validar 40 features (considerando que solo valida lo que desarrollo haya terminado).
Paso 5: Identifica el cuello de botella comparando ambas velocidades.
Paso 6: Indica el tiempo mínimo total del proyecto.

Responde las preguntas:
1. ¿Cuántas semanas mínimo se necesitan?
2. ¿Hay algún cuello de botella? ¿Cuál?
```

#### Respuesta del modelo (con CoT)

### Paso 1: Capacidad semanal de desarrollo

**Frontend**

* 4 desarrolladores × 2 features/semana = 8 features/semana

**Backend**

* 3 desarrolladores × 1.5 features/semana = 4.5 features/semana

**Capacidad total desarrollo**

* 8 + 4.5 = **12.5 features/semana**

---

### Paso 2: Capacidad semanal de QA

* 2 testers × 5 features/semana = **10 features/semana**

---

### Paso 3: Semanas necesarias para desarrollar 40 features

Capacidad desarrollo = 12.5 features/semana

Semanas necesarias:

* 40 ÷ 12.5 = 3.2 semanas

Desarrollo termina en **3.2 semanas**

---

### Paso 4: Semanas necesarias para QA validar 40 features

Capacidad QA = 10 features/semana

Semanas necesarias:

* 40 ÷ 10 = 4 semanas

Pero QA **solo puede validar lo que ya esté desarrollado**, y desarrollo tarda 3.2 semanas.
Como QA es más lento, seguirá validando hasta completar todo.

QA termina en **4 semanas**

---

### Paso 5: Identificar cuello de botella

Comparación:

* Desarrollo: 12.5 features/semana
* QA: 10 features/semana

QA es más lento → **QA es el cuello de botella**

---

### Paso 6: Tiempo mínimo total del proyecto

El proyecto termina cuando:

* Todo esté desarrollado **y**
* Todo esté validado

El más lento es QA:

Tiempo total = **4 semanas**

---

# Respuestas finales

**1. ¿Cuántas semanas mínimo se necesitan?**
→ **4 semanas**

**2. ¿Hay algún cuello de botella? ¿Cuál?**
→ Sí. El cuello de botella es **QA**, porque valida 10 features/semana frente a 12.5 del desarrollo.


### Análisis de diferencias entre ambos enfoques

Ambos resultados son correctos, sin embargo, la respuesta con CoT es bastante más completa y te indica mejor el razonamiento que ha seguido para llegar a esa solución.
