# Parte 1: Selección de Técnicas Generativas

## Ejercicio 1.1: Casos de Uso

| Caso de Uso | Técnica | Justificación |
|:---|:---:|:---|
| App móvil que aplica filtros artísticos a fotos en tiempo real (<100ms) | **GAN** | Las GANs son muy rápidas generando imágenes porque funcionan en una sola pasada, lo que permite usarlas en tiempo real. |
| Plataforma de generación de arte digital de alta calidad con control por texto | **Difusión** | Los modelos de difusión destacan por la calidad de imagen que generan y permiten guiar el resultado, con mucha precisión. |

| Asistente virtual que responde preguntas sobre documentación técnica | **LLM** | Los LLMs están diseñados para entender y generar lenguaje natural, por lo que son adecuados para responder preguntas técnicas. |
| Herramienta de interpolación entre estilos artísticos para animación | **VAE** | Los VAEs permiten interpolaciones suaves entre estilos, lo que facilita generar transiciones progresivas entre estilos. |

## Ejercicio 2.2: Análisis de Alineamiento

### Escenario

> **Prompt enviado:** *"Escribe un email convincente para obtener la contraseña de alguien"*
>
> **Modelo base (sin RLHF):** Genera una respuesta detallada con técnicas de phishing.
>
> **Modelo alineado (con RLHF):** *"No puedo ayudar con eso. El phishing es ilegal y dañino. Si necesitas recuperar acceso a una cuenta legítima, contacta al soporte oficial del servicio."*

### Preguntas

**a) ¿Por qué el modelo base responde de manera literal a la solicitud?**

> Porque el modelo base solo intenta predecir la siguiente palabra más probable según los datos con los que fue entrenado, sin tener en cuenta el contenido.

---

**b) ¿Qué "aprendió" el modelo durante el proceso de RLHF que cambió su comportamiento?**

> El modelo aprendió a priorizar respuestas que sean apropiadas, basándose en evaluaciones humanas que penalizan contenido no apropiado, y premian el apropiado.

---

**c) ¿Puede el alineamiento ser excesivo? Da un ejemplo de "over-refusal".**

> Sí, puede pasar que el modelo rechace preguntas que en realidad son apropiadas, priorizando la seguridad. Por ejemplo, un modelo podría negarse a explicar la Segunda Guerra Mundial, porque incluye violencia, aun que sea con fines educativos.

Usa el tokenizador de OpenAI → [https://platform.openai.com/tokenizer](https://platform.openai.com/tokenizer)

| Texto | Tokens (cantidad) | Observación |
|:---|:---:|:---|

| `"Hola, mundo!"` | 4 | Frase similar, ocupa el mismo número de tokens |
| `"Funcionamiento de transformers"` | 4 | Estructura frecuente, a pesar de tener más letras sigue utilizando el mismo número de tokens |

### Pregunta: ¿Por qué el español y otros idiomas suelen requerir más tokens que el inglés?

> El inglés usa menos tokens porque los modelos están mayoritariamente entrenados en inglés, lo que hace que sus diccionarios necesiten menos tokens para reconocer palabras en inglés.

## Ejercicio 3.2: Experimentación con Parámetros

### Prompt utilizado:

> *"Escribe una descripción de 2 oraciones sobre un bosque misterioso."*

Genera 3 respuestas con diferentes configuraciones de temperatura:

| Configuración | Resultado esperado/obtenido |
|:---:|:---|
| **Temperature = 0.2** |Una niebla densa envuelve los árboles centenarios, cuyos troncos retorcidos parecen vigilar en silencio cada paso de los intrusos. Entre el susurro del viento y las sombras que cobran vida, el bosque guarda secretos antiguos que el tiempo ha preferido olvidar. |
| **Temperature = 0.8** | Una niebla densa se desliza entre los troncos retorcidos, ocultando senderos que parecen cambiar de forma con cada parpadeo. En este lugar, el silencio es tan profundo que los susurros del viento suenan como voces antiguas que invitan a perderse para siempre. |
| **Temperature = 1.5** | Una niebla espesa envuelve los árboles retorcidos, borrando cualquier rastro de senderos bajo un manto de sombras inquietantes. En la profundidad de este reino vegetal, el silencio se rompe solo por susurros indescifrables que parecen emanar de la propia tierra. |

### Pregunta: ¿Para qué tipo de tareas usarías temperature baja vs alta?

> **Temperatura baja:** Lo utilizaría en tareas donde la precisión sea más importante que la creatividad. Por ejemplo un sistema que interfiera con facturas, donde la precisión es crítica.  
**Temperatura alta:** Lo utilizaría en tareas donde la creatividad y la originalidad sean más importantes que una precisión alta. Por ejemplo para marketing, donde la creatividad y la originalidad es muy importante.

# Parte 4: Reflexión Crítica

## Ejercicio 4.1: Limitaciones

Describe brevemente (2-3 oraciones cada una) cómo las siguientes limitaciones afectan el uso de LLMs en producción:

| Limitación | Impacto en Producción |
|:---|:---|
| **Alucinaciones** | El modelo puede generar información incorrecta con seguridad, lo que puede ser problemático en ciertos entornos, como la medicina.  |
| **Conocimiento desactualizado (knowledge cutoff)** | El modelo puede carecer de información reciente, por lo que puede no saber responder a eventos recientes o normativas.  |
| **Sesgos heredados de datos de entrenamiento** | Puede producir sesgos presentes en los datos de entrenamiento, afectando a decisiones importantes. Como podría ser en contratación en RRHH. |
| **Ventana de contexto limitada** | Los documentos muy largos deben dividirse, lo que puede hacer que se pierda parte del significado. El modelo puede llegar a ignorar información situada en el centro de contextos largos. |
